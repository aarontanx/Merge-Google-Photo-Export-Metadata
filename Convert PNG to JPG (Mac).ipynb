##### Check Folder Structure, Make sure it's in the month format

```python
folder_to_check = [x[0] for x in os.walk(month_folder)][1:]
folder_to_check
```

#### Check each file from each folder
1. If file ends with .png, convert to .jpg
2. If file ends with png.json, convert to jpg.json, replace if same
3. Run exiftool, combine metadata
4. Delete every single file ends with .json

End result would be each file remain in the same folder structure, but all json file will be deleted

```python
file_counter = 0
for j in folder_to_check[:]: #First Level Folder - Find Each Folder
    png_jpg_convert_counter = 0
    json_delete_counter = 0
    jpg_json_convert_counter = 0
    print("Folder "+j)

    for i in os.listdir(j): #Second Level Folder - Find Each File
        if str.lower(i[-4:]) == '.png': #If File ends with .png, convert to .jpg
            im1 = Image.open(os.path.join(j, i))
            im1 = im1.convert('RGB')
            im1.save(os.path.join(j, i[:-4]+".jpg"))
            os.remove(os.path.join(j, i ))
            png_jpg_convert_counter += 1
        if str.lower(i[-8:]) == 'png.json': #If File ends with png.json, convert to jpg.json
            move(os.path.join(j, i), \
                os.path.join(j, i[:-8]+"jpg.json"))
            jpg_json_convert_counter += 1

        #Run Exiftool to fix metadata in this folder
    !exiftool --ext json -r -tagsfromfile "%d/%F.json" "-GPSAltitude<GeoDataAltitude" "-GPSLatitude<GeoDataLatitude" "-GPSLatitudeRef<GeoDataLatitude" "-GPSLongitude<GeoDataLongitude" "-GPSLongitudeRef<GeoDataLongitude" -Description -d %s "-Alldates<PhotoTakenTimeTimestamp" -overwrite_original $j
        
    for i in os.listdir(j): #Loop again to delete json file this time
        if str.lower(i[-4:]) == 'json':
            os.remove(os.path.join(j,i)) #Delete json file     
            json_delete_counter += 1
        else:
            file_counter +=1
                
    print("PNG to JPG Convert Count = "+ str(png_jpg_convert_counter))
    print("JPG JSON Convert Count = "+ str(jpg_json_convert_counter))
    print("Json Delete Count = "+ str(json_delete_counter))
    print("")
print("Metadata Update DONE! Total File: "+str(file_counter))
```

##### Move all the files into one combined folder, name them with number as prefix for each file type

```python
with open("output_move.txt", "a") as f:
    jpg_counter = 1
    mov_counter = 1
    mp4_counter = 1
    other_counter = 1
    heic_counter = 1
    gif_counter = 1

    for j in folder_to_check: #First Level Folder - Find Each Folder
        for i in os.listdir(j): #Second Level Folder - Find Each File
            if str.lower(i[-4:]) == '.jpg':
                shutil.move(os.path.join(j, i), os.path.join(combined_folder, str(jpg_counter).zfill(5)+"_"+i[:-4]+".JPG"))
                jpg_counter += 1
            elif str.lower(i[-5:]) == '.jpeg':
                shutil.move(os.path.join(j, i), os.path.join(combined_folder, str(jpg_counter).zfill(5)+"_"+i[:-5]+".JPG"))
                jpg_counter += 1
            elif str.lower(i[-4:]) == '.mp4':
                shutil.move(os.path.join(j, i), os.path.join(combined_folder, str(mp4_counter).zfill(5)+"_"+i[:-4]+".mp4"))
                mp4_counter += 1
            elif str.lower(i[-4:]) == '.mov':
                shutil.move(os.path.join(j, i), os.path.join(combined_folder, str(mov_counter).zfill(5)+"_"+i[:-4]+".mov"))
                mov_counter += 1
            elif str.lower(i[-5:]) == '.heic':
                shutil.move(os.path.join(j, i), os.path.join(combined_folder, str(heic_counter).zfill(5)+"_"+i[:-5]+".HEIC"))
                heic_counter += 1
            elif str.lower(i[-4:]) == '.gif':
                shutil.move(os.path.join(j, i), os.path.join(combined_folder, str(gif_counter).zfill(5)+"_"+i[:-4]+".gif"))
                gif_counter += 1
            else:
                shutil.move(os.path.join(j, i), os.path.join(combined_folder, str(other_counter).zfill(5)+"_"+i+".other"))
                other_counter += 1

    print("\n", file=f)
    print("JPG Count: "+str(jpg_counter-1), file=f)
    print("MOV Count: "+str(mov_counter-1), file=f)
    print("MP4 Count: "+str(mp4_counter-1), file=f)
    print("HEIC Count: "+str(heic_counter-1), file=f)
    print("GIF Count: "+str(gif_counter-1), file=f)
    print("Other Count: "+str(other_counter-1), file=f)
```

##### Split Files from Combined to Each Folder Category

```python
for i in os.listdir(combined_folder):
    if i != '.DS_Store':
        if str.lower(i[-4:]) == '.jpg':
            shutil.move(os.path.join(combined_folder, i), os.path.join(jpg_folder,i))
        elif str.lower(i[-5:]) == '.heic':
            shutil.move(os.path.join(combined_folder, i), os.path.join(HEIC_folder,i))
        elif str.lower(i[-4:]) == '.mp4':
            shutil.move(os.path.join(combined_folder, i), os.path.join(mp4_folder,i))
        elif str.lower(i[-4:]) == '.mov':
            shutil.move(os.path.join(combined_folder, i), os.path.join(mov_folder,i))
        elif str.lower(i[-4:]) == '.gif':
            shutil.move(os.path.join(combined_folder, i), os.path.join(gif_folder,i))
        else:
            shutil.move(os.path.join(combined_folder, i), os.path.join(other_folder,i))

print("Done!")
        
```
