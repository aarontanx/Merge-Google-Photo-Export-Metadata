```python
with open("output.txt", "a") as f:
    file_counter = 0
    for j in folder_to_check[(570+985):]: #First Level Folder - Find Each Folder
        png_jpg_convert_counter = 0
        json_delete_counter = 0
        jpg_json_convert_counter = 0
        print("Folder "+j, file=f)
        print("Folder "+j)

        for i in os.listdir(j): #Second Level Folder - Find Each File

            if str.lower(i[-4:]) == '.png': #If File ends with .png, convert to .jpg
                im1 = Image.open(os.path.join(j, i))
                im1 = im1.convert('RGB')
                im1.save(os.path.join(j, i[:-4]+".jpg"))
                os.remove(os.path.join(j, i ))
                png_jpg_convert_counter += 1
            if str.lower(i[-8:]) == 'png.json': #If File ends with png.json, convert to jpg.json
                move(j+"\\"+i, j+"\\"+i[:-8]+"jpg.json")
                jpg_json_convert_counter += 1


        #Run Exiftool to fix metadata in this folder
        x = !exiftool --ext json -r -tagsfromfile "%d/%F.json" "-GPSAltitude<GeoDataAltitude" "-GPSLatitude<GeoDataLatitude" "-GPSLatitudeRef<GeoDataLatitude" "-GPSLongitude<GeoDataLongitude" "-GPSLongitudeRef<GeoDataLongitude" -Description -d %s "-Alldates<PhotoTakenTimeTimestamp" -overwrite_original $j
        
        print(x, file=f)
        
        
        for i in os.listdir(j): #Loop again to delete json file this time
            if str.lower(i[-4:]) == 'json':
                os.remove(j+"\\"+i) #Delete json file
                json_delete_counter += 1
            else:
                file_counter +=1

        print("PNG to JPG Convert Count = "+ str(png_jpg_convert_counter), file=f)
        print("JPG JSON Convert Count = "+ str(jpg_json_convert_counter), file=f)
        print("Json Delete Count = "+ str(json_delete_counter), file=f)
        print("\n", file=f)
        print("PNG to JPG Convert Count = "+ str(png_jpg_convert_counter))
        print("JPG JSON Convert Count = "+ str(jpg_json_convert_counter))
        print("Json Delete Count = "+ str(json_delete_counter))
        print("\n")
    print("Metadata Update DONE! Total File: "+str(file_counter), file=f)
```

```python
with open("output_move.txt", "a") as f:
    jpg_counter = 1
    mov_counter = 1
    mp4_counter = 1
    other_counter = 1
    heic_counter = 1
    gif_counter = 1

    for j in folder_to_check: #First Level Folder - Find Each Folder
        for i in os.listdir(j): #Second Level Folder - Find Each File
            if str.lower(i[-4:]) == '.jpg':
                shutil.move(os.path.join(j, i), os.path.join(compile_folder, str(jpg_counter).zfill(5)+"_"+i[:-4]+".JPG"))
                jpg_counter += 1
            elif str.lower(i[-5:]) == '.jpeg':
                shutil.move(os.path.join(j, i), os.path.join(compile_folder, str(jpg_counter).zfill(5)+"_"+i[:-5]+".JPG"))
                jpg_counter += 1
            elif str.lower(i[-4:]) == '.mp4':
                shutil.move(os.path.join(j, i), os.path.join(compile_folder, str(mp4_counter).zfill(5)+"_"+i[:-4]+".mp4"))
                mp4_counter += 1
            elif str.lower(i[-4:]) == '.mov':
                shutil.move(os.path.join(j, i), os.path.join(compile_folder, str(mov_counter).zfill(5)+"_"+i[:-4]+".mov"))
                mov_counter += 1
            elif str.lower(i[-5:]) == '.heic':
                shutil.move(os.path.join(j, i), os.path.join(compile_folder, str(mov_counter).zfill(5)+"_"+i[:-5]+".HEIC"))
                heic_counter += 1
            elif str.lower(i[-4:]) == '.gif':
                shutil.move(os.path.join(j, i), os.path.join(compile_folder, str(mov_counter).zfill(5)+"_"+i[:-4]+".gif"))
                mov_counter += 1
            else:
                shutil.move(os.path.join(j, i), os.path.join(compile_folder, str(other_counter).zfill(5)+"_"+i+".other"))
                other_counter += 1

    print("\n", file=f)
    print("JPG Count: "+str(jpg_counter-1), file=f)
    print("MOV Count: "+str(mov_counter-1), file=f)
    print("MPV Count: "+str(mp4_counter-1), file=f)
    print("HEIC Count: "+str(heic_counter-1), file=f)
    print("GIF Count: "+str(gif_counter-1), file=f)
    print("Other Count: "+str(other_counter-1), file=f)
```

```python
jpg_folder = os.path.join(parent_path, "Separated_Folder", "jpg_folder")
HEIC_folder = os.path.join(parent_path, "Separated_Folder", "HEIC_folder")
mp4_folder = os.path.join(parent_path, "Separated_Folder", "mp4_folder")
mov_folder = os.path.join(parent_path, "Separated_Folder", "mov_folder")
gif_folder = os.path.join(parent_path, "Separated_Folder", "gif_folder")
other_folder = os.path.join(parent_path, "Separated_Folder", "other_folder")
```

```python
for i in os.listdir(compile_folder):
    if str.lower(i[-4:]) == '.jpg':
        shutil.move(os.path.join(compile_folder, i), os.path.join(jpg_folder))
    elif str.lower(i[-5:]) == '.heic':
        shutil.move(os.path.join(compile_folder, i), os.path.join(HEIC_folder))
    elif str.lower(i[-4:]) == '.mp4':
        shutil.move(os.path.join(compile_folder, i), os.path.join(mp4_folder))
    elif str.lower(i[-4:]) == '.mov':
        shutil.move(os.path.join(compile_folder, i), os.path.join(mov_folder))
    else:
        shutil.move(os.path.join(compile_folder, i), os.path.join(other_folder))

print("Done!")
        
```
